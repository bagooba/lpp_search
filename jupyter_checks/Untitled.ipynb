```python
def keep_unique_period_indices(periods, merge_tol=0.02):
    kept = []

    for i, Pi in enumerate(periods):
        keep = True

        for j in kept:
            Pj = periods[j]

            if abs(Pi - Pj) / Pj < merge_tol:
                keep = False
                break

        if keep:
            kept.append(i)

    return kept

def potential_periods_from_single_dt_events(t0s, min_support=3, perc_tol = 0.02, max_ratio = 3, P_min = 1):
#     t0s = [ev.t0_days for ev in events]
    
    potential_periods = np.unique(np.abs([x-y for x in t0s for y in t0s]))[1:]
    
    ratios = np.array([potential_periods/x for x in potential_periods])
#     print(ratios)
    all_checks  = np.array([ratios*x for x  in range(1, max_ratio+1)])
    all_checks[all_checks<=1] = np.nan
    
    all_periods = []
    matches = np.zeros(ratios.shape, dtype = bool)
    for iii, ary_check in enumerate(all_checks):
        percent_diff = np.abs((ary_check - np.round(ary_check))/ary_check)
        close = percent_diff<perc_tol
        check_close = np.nansum(close, axis = 1)>=min_support
            
        all_periods.extend(list(potential_periods[check_close]/(iii+1)))
#     all_periods = np.array([x for x in ary for ary in all_periods])
#     print(all_periods)
    all_periods_indxs = keep_unique_period_indices(all_periods)
    final_periods     = np.unique(np.array(all_periods)[all_periods_indxs])
    return final_periods[final_periods>P_min]
```
